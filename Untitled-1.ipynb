```python
import pyodbc
import pandas as pd

```

```python
conn_str = r'DRIVER={Microsoft Access Driver (*.mdb, *.accdb)};DBQ=C:\Users\gia.luongdo\Desktop\ERP-Import.mdb;'
conn = pyodbc.connect(conn_str)

```

```python
df = pd.read_sql("SELECT * FROM t_Art_MegaBase", conn)

```

```python
# Remove duplicate rows
df_cleaned = df.drop_duplicates()

# Strip whitespace from string columns
str_cols = df_cleaned.select_dtypes(include='object').columns
df_cleaned[str_cols] = df_cleaned[str_cols].apply(lambda x: x.str.strip())

# Replace empty strings with NaN
df_cleaned[str_cols] = df_cleaned[str_cols].replace('', pd.NA)

# Optionally, drop rows with all NaN values
df_cleaned = df_cleaned.dropna(how='all')

# Show summary of cleaned data
print(df_cleaned.info())
```
